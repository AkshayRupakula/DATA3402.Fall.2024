1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter.

```python
class Counter:
    def __init__(self, max_value):
        """
        Initialize the counter with a specified maximum value.
        """
        self.max_value = max_value
        self.current_value = 0

    def increment(self):
        """
        Increment the counter by 1, but print an error message if the counter
        exceeds the specified maximum value.
        """
        if self.current_value < self.max_value:
            self.current_value += 1
        else:
            print(f"Error: Counter cannot exceed the maximum value of {self.max_value}.")

    def reset(self):
        """
        Reset the counter to 0.
        """
        self.current_value = 0

    def get_value(self):
        """
        Get the current value of the counter.
        """
        return self.current_value

# Usage:
counter = Counter(5)  # Set the maximum value to 5

# Increment the counter a few times
for _ in range(7):
    counter.increment()
    print(counter.get_value())  # Print the current value

# Reset the counter
counter.reset()
print("Counter after reset:", counter.get_value())
```

3. Implement a class to represent a rectangle, holding the length, width, and $x$ and $y$ coordinates of a corner of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members.

```python
class Rectangle:
    def __init__(self, length, width, x, y):
        """
        Initialize the rectangle with length, width, and the (x, y) coordinates of one corner.
        """
        self.__length = length  # Private length
        self.__width = width    # Private width
        self.__x = x            # Private x-coordinate
        self.__y = y            # Private y-coordinate

    def area(self):
        """
        Compute and return the area of the rectangle.
        """
        return self.__length * self.__width

    def perimeter(self):
        """
        Compute and return the perimeter of the rectangle.
        """
        return 2 * (self.__length + self.__width)

    # Accessor methods
    def get_length(self):
        """
        Get the length of the rectangle.
        """
        return self.__length

    def get_width(self):
        """
        Get the width of the rectangle.
        """
        return self.__width

    def get_x(self):
        """
        Get the x-coordinate of the rectangle's corner.
        """
        return self.__x

    def get_y(self):
        """
        Get the y-coordinate of the rectangle's corner.
        """
        return self.__y

# Usage:
rectangle = Rectangle(10, 5, 0, 0)  # Create a rectangle with length=10, width=5, and corner at (0, 0)

# Calculate and print area and perimeter
print(f"Area: {rectangle.area()}")  # Should output 50
print(f"Perimeter: {rectangle.perimeter()}")  # Should output 30

# Access and print private data members using accessor methods
print(f"Length: {rectangle.get_length()}")
print(f"Width: {rectangle.get_width()}")
print(f"X-coordinate: {rectangle.get_x()}")
print(f"Y-coordinate: {rectangle.get_y()}")
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
import math
from abc import ABC, abstractmethod

# Define the abstract base class 'Shape'
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# Now we can define the Triangle class inheriting from Shape
class Triangle(Shape):
    def __init__(self, a, b, c, x, y):
        """
        Initialize the triangle with sides a, b, and c, and the (x, y) coordinates of one vertex.
        """
        self.__a = a  # Side a
        self.__b = b  # Side b
        self.__c = c  # Side c
        self.__x = x  # x-coordinate of one vertex
        self.__y = y  # y-coordinate of one vertex

    def area(self):
        """
        Compute and return the area of the triangle using Heron's formula.
        """
        s = (self.__a + self.__b + self.__c) / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def perimeter(self):
        """
        Compute and return the perimeter of the triangle.
        """
        return self.__a + self.__b + self.__c

    def get_x(self):
        """
        Get the x-coordinate of the triangle's vertex.
        """
        return self.__x

    def get_y(self):
        """
        Get the y-coordinate of the triangle's vertex.
        """
        return self.__y

    def get_sides(self):
        """
        Get the sides of the triangle.
        """
        return self.__a, self.__b, self.__c


# Usage:
triangle = Triangle(3, 4, 5, 0, 0) 

print(f"Triangle Area: {triangle.area():.2f}")  
print(f"Triangle Perimeter: {triangle.perimeter():.2f}")  

print(f"X-coordinate of vertex: {triangle.get_x()}")
print(f"Y-coordinate of vertex: {triangle.get_y()}")
print(f"Sides of triangle: {triangle.get_sides()}")
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object.

```python
import math
from abc import ABC, abstractmethod

# Base class for Shape
class Shape(ABC):
    """
    Abstract base class to represent a generic shape.
    """

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    @abstractmethod
    def get_x(self):
        pass

    @abstractmethod
    def get_y(self):
        pass

    @abstractmethod
    def get_perimeter_points(self):
        """
        Abstract method to return a list of up to 16 (x, y) points on the perimeter of the shape.
        """
        pass

# Rectangle inherits from Shape
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter_points(self):
        """
        Returns up to 16 points on the perimeter of the rectangle.
        """
        points = []
        # Define the four corners
        points.append((self.__x, self.__y))  # Bottom-left corner
        points.append((self.__x + self.__length, self.__y))  # Bottom-right corner
        points.append((self.__x + self.__length, self.__y + self.__width))  # Top-right corner
        points.append((self.__x, self.__y + self.__width))  # Top-left corner

        # Add intermediate points along the edges
        if self.__length > 1 and self.__width > 1:
            step_x = self.__length / 4
            step_y = self.__width / 4
            for i in range(1, 4):
                points.append((self.__x + i * step_x, self.__y))  # Bottom edge
                points.append((self.__x + i * step_x, self.__y + self.__width))  # Top edge
                points.append((self.__x, self.__y + i * step_y))  # Left edge
                points.append((self.__x + self.__length, self.__y + i * step_y))  # Right edge

        return points[:16]  # Return at most 16 points

# Circle class inheriting from Shape
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def area(self):
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter_points(self):
        """
        Returns up to 16 points on the perimeter of the circle.
        """
        points = []
        num_points = min(16, int(self.__radius * math.pi))  # Limit to 16 points

        for i in range(num_points):
            angle = (2 * math.pi / num_points) * i
            x_point = self.__x + self.__radius * math.cos(angle)
            y_point = self.__y + self.__radius * math.sin(angle)
            points.append((x_point, y_point))

        return points

# Triangle class inheriting from Shape
class Triangle(Shape):
    def __init__(self, a, b, c, x, y):
        self.__a = a
        self.__b = b
        self.__c = c
        self.__x = x
        self.__y = y

    def area(self):
        s = (self.__a + self.__b + self.__c) / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def perimeter(self):
        return self.__a + self.__b + self.__c

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter_points(self):
        """
        Returns up to 16 points on the perimeter of the triangle.
        """
        points = []
        points.append((self.__x, self.__y))  # First vertex
        points.append((self.__x + self.__a, self.__y))  # Second vertex (along a-axis)
        points.append((self.__x + self.__a / 2, self.__y + math.sqrt(self.__b ** 2 - (self.__a / 2) ** 2)))  # Third vertex

        # Add intermediate points along the sides
        if self.__a > 1 and self.__b > 1 and self.__c > 1:
            num_points_per_side = min(5, 16 // 3) 
            for i in range(1, num_points_per_side):
                points.append((self.__x + i * (self.__a / num_points_per_side), self.__y))  # Side 1

        return points[:16]

# Example usage
rectangle = Rectangle(10, 5, 0, 0)  # Rectangle with length 10, width 5, bottom-left at (0, 0)
circle = Circle(7, 3, 3)  # Circle with radius 7, center at (3, 3)
triangle = Triangle(3, 4, 5, 0, 0)  # Triangle with sides 3, 4, 5, first vertex at (0, 0)

# Get perimeter points for each shape
rect_points = rectangle.get_perimeter_points()
circle_points = circle.get_perimeter_points()
triangle_points = triangle.get_perimeter_points()

print(f"Rectangle Perimeter Points: {rect_points}")
print(f"Circle Perimeter Points: {circle_points}")
print(f"Triangle Perimeter Points: {triangle_points}")
```

```python
import math
from abc import ABC, abstractmethod

# Base class for Shape
class Shape(ABC):
    """
    Abstract base class to represent a generic shape.
    """

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    @abstractmethod
    def get_x(self):
        pass

    @abstractmethod
    def get_y(self):
        pass

    @abstractmethod
    def get_perimeter_points(self):
        pass

    @abstractmethod
    def is_point_inside(self, x, y):
        """
        Abstract method to test if a point (x, y) is inside the shape.
        """
        pass

# Rectangle class inheriting from Shape
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter_points(self):
        points = []
        points.append((self.__x, self.__y))  # Bottom-left corner
        points.append((self.__x + self.__length, self.__y))  # Bottom-right corner
        points.append((self.__x + self.__length, self.__y + self.__width))  # Top-right corner
        points.append((self.__x, self.__y + self.__width))  # Top-left corner
        return points[:16]

    def is_point_inside(self, x, y):
        """
        Test if a point (x, y) is inside the rectangle.
        """
        return self.__x <= x <= (self.__x + self.__length) and self.__y <= y <= (self.__y + self.__width)

# Circle class inheriting from Shape
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x  # x-coordinate of center
        self.__y = y  # y-coordinate of center

    def area(self):
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter_points(self):
        points = []
        num_points = min(16, int(self.__radius * math.pi))  # Limit to 16 points
        for i in range(num_points):
            angle = (2 * math.pi / num_points) * i
            x_point = self.__x + self.__radius * math.cos(angle)
            y_point = self.__y + self.__radius * math.sin(angle)
            points.append((x_point, y_point))
        return points

    def is_point_inside(self, x, y):
        """
        Test if a point (x, y) is inside the circle.
        """
        distance = math.sqrt((x - self.__x) ** 2 + (y - self.__y) ** 2)
        return distance <= self.__radius

# Triangle class inheriting from Shape
class Triangle(Shape):
    def __init__(self, a, b, c, x, y):
        self.__a = a
        self.__b = b
        self.__c = c
        self.__x = x
        self.__y = y

    def area(self):
        s = (self.__a + self.__b + self.__c) / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def perimeter(self):
        return self.__a + self.__b + self.__c

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter_points(self):
        points = []
        points.append((self.__x, self.__y))  # First vertex
        points.append((self.__x + self.__a, self.__y))  # Second vertex (along a-axis)
        points.append((self.__x + self.__a / 2, self.__y + math.sqrt(self.__b ** 2 - (self.__a / 2) ** 2)))  # Third vertex
        return points[:16]

    def is_point_inside(self, x, y):
        """
        Test if a point (x, y) is inside the triangle using the area method.
        """
        # Vertices of the triangle
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__a, self.__y
        x3, y3 = self.__x + self.__a / 2, self.__y + math.sqrt(self.__b ** 2 - (self.__a / 2) ** 2)

        # Calculate area of the full triangle
        def triangle_area(x1, y1, x2, y2, x3, y3):
            return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2.0)

        A = triangle_area(x1, y1, x2, y2, x3, y3)

        # Calculate areas of the sub-triangles formed with the point
        A1 = triangle_area(x, y, x2, y2, x3, y3)
        A2 = triangle_area(x1, y1, x, y, x3, y3)
        A3 = triangle_area(x1, y1, x2, y2, x, y)

        # Check if sum of A1, A2, and A3 is equal to A
        return A == A1 + A2 + A3

# Example usage
rectangle = Rectangle(10, 5, 0, 0)  # Rectangle with length 10, width 5, bottom-left at (0, 0)
circle = Circle(7, 3, 3)  # Circle with radius 7, center at (3, 3)
triangle = Triangle(3, 4, 5, 0, 0)  # Triangle with sides 3, 4, 5, first vertex at (0, 0)

# Test point inside shapes
print(f"Is point (5, 3) inside rectangle? {rectangle.is_point_inside(5, 3)}")  # True
print(f"Is point (11, 5) inside rectangle? {rectangle.is_point_inside(11, 5)}")  # False

print(f"Is point (3, 3) inside circle? {circle.is_point_inside(3, 3)}")  # True
print(f"Is point (10, 10) inside circle? {circle.is_point_inside(10, 10)}")  # False

print(f"Is point (1, 1) inside triangle? {triangle.is_point_inside(1, 1)}")  # True
print(f"Is point (6, 2) inside triangle? {triangle.is_point_inside(6, 2)}")  # False
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
import math
from abc import ABC, abstractmethod

class Shape(ABC):
    """
    Abstract base class to represent a generic shape.
    """

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    @abstractmethod
    def get_x(self):
        pass

    @abstractmethod
    def get_y(self):
        pass

    @abstractmethod
    def get_perimeter_points(self):
        pass

    @abstractmethod
    def is_point_inside(self, x, y):
        pass

    @abstractmethod
    def overlaps(self, other_shape):
        """
        Abstract method to check if the current shape overlaps with another shape.
        """
        pass


# Rectangle class inheriting from Shape
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter_points(self):
        points = []
        points.append((self.__x, self.__y))  # Bottom-left corner
        points.append((self.__x + self.__length, self.__y))  # Bottom-right corner
        points.append((self.__x + self.__length, self.__y + self.__width))  # Top-right corner
        points.append((self.__x, self.__y + self.__width))  # Top-left corner
        return points[:16]

    def is_point_inside(self, x, y):
        """
        Test if a point (x, y) is inside the rectangle.
        """
        return self.__x <= x <= (self.__x + self.__length) and self.__y <= y <= (self.__y + self.__width)

    def overlaps(self, other_shape):
        """
        Test if the rectangle overlaps with another shape.
        For now, we test for rectangle-rectangle and rectangle-circle overlaps.
        """
        if isinstance(other_shape, Rectangle):
            # Rectangle-Rectangle overlap test
            return not (self.__x + self.__length < other_shape.get_x() or
                        other_shape.get_x() + other_shape.__length < self.__x or
                        self.__y + self.__width < other_shape.get_y() or
                        other_shape.get_y() + other_shape.__width < self.__y)

        if isinstance(other_shape, Circle):
            # Rectangle-Circle overlap test
            circle_x, circle_y = other_shape.get_x(), other_shape.get_y()
            circle_radius = other_shape.get_radius()

            # Find the closest point on the rectangle to the circle's center
            closest_x = max(self.__x, min(circle_x, self.__x + self.__length))
            closest_y = max(self.__y, min(circle_y, self.__y + self.__width))

            # Check if the closest point is within the circle's radius
            distance = math.sqrt((closest_x - circle_x) ** 2 + (closest_y - circle_y) ** 2)
            return distance <= circle_radius

        return False


# Circle class inheriting from Shape
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x  # x-coordinate of center
        self.__y = y  # y-coordinate of center

    def area(self):
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_radius(self):
        return self.__radius

    def get_perimeter_points(self):
        points = []
        num_points = min(16, int(self.__radius * math.pi))  # Limit to 16 points
        for i in range(num_points):
            angle = (2 * math.pi / num_points) * i
            x_point = self.__x + self.__radius * math.cos(angle)
            y_point = self.__y + self.__radius * math.sin(angle)
            points.append((x_point, y_point))
        return points

    def is_point_inside(self, x, y):
        """
        Test if a point (x, y) is inside the circle.
        """
        distance = math.sqrt((x - self.__x) ** 2 + (y - self.__y) ** 2)
        return distance <= self.__radius

    def overlaps(self, other_shape):
        """
        Test if the circle overlaps with another shape.
        For now, we test for circle-circle and circle-rectangle overlaps.
        """
        if isinstance(other_shape, Circle):
            # Circle-Circle overlap test
            distance = math.sqrt((self.__x - other_shape.get_x()) ** 2 + (self.__y - other_shape.get_y()) ** 2)
            return distance <= (self.__radius + other_shape.get_radius())

        if isinstance(other_shape, Rectangle):
            # Circle-Rectangle overlap (this is handled in the Rectangle class)
            return other_shape.overlaps(self)

        return False


# Triangle class inheriting from Shape
class Triangle(Shape):
    def __init__(self, a, b, c, x, y):
        self.__a = a
        self.__b = b
        self.__c = c
        self.__x = x
        self.__y = y

    def area(self):
        s = (self.__a + self.__b + self.__c) / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def perimeter(self):
        return self.__a + self.__b + self.__c

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter_points(self):
        points = []
        points.append((self.__x, self.__y))  # First vertex
        points.append((self.__x + self.__a, self.__y))  # Second vertex (along a-axis)
        points.append((self.__x + self.__a / 2, self.__y + math.sqrt(self.__b ** 2 - (self.__a / 2) ** 2)))  # Third vertex
        return points[:16]

    def is_point_inside(self, x, y):
        """
        Test if a point (x, y) is inside the triangle using the area method.
        """
        # Vertices of the triangle
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__a, self.__y
        x3, y3 = self.__x + self.__a / 2, self.__y + math.sqrt(self.__b ** 2 - (self.__a / 2) ** 2)

        # Calculate area of the full triangle
        def triangle_area(x1, y1, x2, y2, x3, y3):
            return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2.0)

        A = triangle_area(x1, y1, x2, y2, x3, y3)

        # Calculate areas of the sub-triangles formed with the point
        A1 = triangle_area(x, y, x2, y2, x3, y3)
        A2 = triangle_area(x1, y1, x, y, x3, y3)
        A3 = triangle_area(x1, y1, x2, y2, x, y)

        # Check if sum of A1, A2, and A3 is equal to A
        return A == A1 + A2 + A3

    def overlaps(self, other_shape):
        """
        For now, we handle overlaps with rectangles or circles.
        """
        # Simplified check for Triangle overlap
        for point in self.get_perimeter_points():
            if other_shape.is_point_inside(point[0], point[1]):
                return True

        for point in other_shape.get_perimeter_points():
            if self.is_point_inside(point[0], point[1]):
                return True

        return False

# Example usage
rectangle1 = Rectangle(10, 5, 0, 0)
rectangle2 = Rectangle(3, 3, 5, 2)  # Overlapping rectangles

circle1 = Circle(5, 2, 2)
circle2 = Circle(3, 7, 7)  # Non-overlapping circles

# Test rectangle overlap
print(f"Do the rectangles overlap? {rectangle1.overlaps(rectangle2)}")  # True

# Test circle overlap
print(f"Do the circles overlap? {circle1.overlaps(circle2)}")  # False

# Test rectangle-circle overlap
print(f"Does rectangle1 overlap with circle1? {rectangle1.overlaps(circle1)}")  # True
```

```python
import math
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from abc import ABC, abstractmethod

# Base class for Shape
class Shape(ABC):
    """
    Abstract base class to represent a generic shape.
    """

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    @abstractmethod
    def get_x(self):
        pass

    @abstractmethod
    def get_y(self):
        pass

    @abstractmethod
    def get_perimeter_points(self):
        pass

    @abstractmethod
    def is_point_inside(self, x, y):
        pass

    @abstractmethod
    def overlaps(self, other_shape):
        pass

    @abstractmethod
    def paint(self, ax):
        pass

# Rectangle class inheriting from Shape
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def __repr__(self):
        return f"Rectangle({self.__length}, {self.__width}, {self.__x}, {self.__y})"

    def get_perimeter_points(self):
        points = []
        points.append((self.__x, self.__y))  # Bottom-left corner
        points.append((self.__x + self.__length, self.__y))  # Bottom-right corner
        points.append((self.__x + self.__length, self.__y + self.__width))  # Top-right corner
        points.append((self.__x, self.__y + self.__width))  # Top-left corner
        return points[:16]

    def is_point_inside(self, x, y):
        return self.__x <= x <= (self.__x + self.__length) and self.__y <= y <= (self.__y + self.__width)

    def overlaps(self, other_shape):
        if isinstance(other_shape, Rectangle):
            return not (self.__x + self.__length < other_shape.get_x() or
                        other_shape.get_x() + other_shape.__length < self.__x or
                        self.__y + self.__width < other_shape.get_y() or
                        other_shape.get_y() + other_shape.__width < self.__y)

        if isinstance(other_shape, Circle):
            circle_x, circle_y = other_shape.get_x(), other_shape.get_y()
            circle_radius = other_shape.get_radius()
            closest_x = max(self.__x, min(circle_x, self.__x + self.__length))
            closest_y = max(self.__y, min(circle_y, self.__y + self.__width))
            distance = math.sqrt((closest_x - circle_x) ** 2 + (closest_y - circle_y) ** 2)
            return distance <= circle_radius

        return False

    def paint(self, ax):
        rectangle = patches.Rectangle((self.__x, self.__y), self.__length, self.__width, linewidth=1, edgecolor='blue', facecolor='none')
        ax.add_patch(rectangle)


# Circle class inheriting from Shape
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def area(self):
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_radius(self):
        return self.__radius

    def __repr__(self):
        return f"Circle({self.__radius}, {self.__x}, {self.__y})"

    def get_perimeter_points(self):
        points = []
        num_points = min(16, int(self.__radius * math.pi))
        for i in range(num_points):
            angle = (2 * math.pi / num_points) * i
            x_point = self.__x + self.__radius * math.cos(angle)
            y_point = self.__y + self.__radius * math.sin(angle)
            points.append((x_point, y_point))
        return points

    def is_point_inside(self, x, y):
        distance = math.sqrt((x - self.__x) ** 2 + (y - self.__y) ** 2)
        return distance <= self.__radius

    def overlaps(self, other_shape):
        if isinstance(other_shape, Circle):
            distance = math.sqrt((self.__x - other_shape.get_x()) ** 2 + (self.__y - other_shape.get_y()) ** 2)
            return distance <= (self.__radius + other_shape.get_radius())

        if isinstance(other_shape, Rectangle):
            return other_shape.overlaps(self)

        return False

    def paint(self, ax):
        circle = patches.Circle((self.__x, self.__y), self.__radius, linewidth=1, edgecolor='red', facecolor='none')
        ax.add_patch(circle)


# Triangle class inheriting from Shape
class Triangle(Shape):
    def __init__(self, a, b, c, x, y):
        self.__a = a
        self.__b = b
        self.__c = c
        self.__x = x
        self.__y = y

    def area(self):
        s = (self.__a + self.__b + self.__c) / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def perimeter(self):
        return self.__a + self.__b + self.__c

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def __repr__(self):
        return f"Triangle({self.__a}, {self.__b}, {self.__c}, {self.__x}, {self.__y})"

    def get_perimeter_points(self):
        points = []
        points.append((self.__x, self.__y))
        points.append((self.__x + self.__a, self.__y))
        points.append((self.__x + self.__a / 2, self.__y + math.sqrt(self.__b ** 2 - (self.__a / 2) ** 2)))
        return points[:16]

    def is_point_inside(self, x, y):
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__a, self.__y
        x3, y3 = self.__x + self.__a / 2, self.__y + math.sqrt(self.__b ** 2 - (self.__a / 2) ** 2)
        def triangle_area(x1, y1, x2, y2, x3, y3):
            return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2.0)
        A = triangle_area(x1, y1, x2, y2, x3, y3)
        A1 = triangle_area(x, y, x2, y2, x3, y3)
        A2 = triangle_area(x1, y1, x, y, x3, y3)
        A3 = triangle_area(x1, y1, x2, y2, x, y)
        return A == A1 + A2 + A3

    def overlaps(self, other_shape):
        for point in self.get_perimeter_points():
            if other_shape.is_point_inside(point[0], point[1]):
                return True
        for point in other_shape.get_perimeter_points():
            if self.is_point_inside(point[0], point[1]):
                return True
        return False

    def paint(self, ax):
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__a, self.__y
        x3, y3 = self.__x + self.__a / 2, self.__y + math.sqrt(self.__b ** 2 - (self.__a / 2) ** 2)
        polygon = patches.Polygon([(x1, y1), (x2, y2), (x3, y3)], linewidth=1, edgecolor='green', facecolor='none')
        ax.add_patch(polygon)

class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def save(self, filename):
        with open(filename, "w") as file:
            for shape in self.shapes:
                file.write(repr(shape) + "\n")

    @classmethod
    def load(cls, filename):
        drawing = cls()
        with open(filename, "r") as file:
            for line in file:
                drawing.add_shape(eval(line.strip()))
        return drawing

    def paint(self):
        fig, ax = plt.subplots()
        ax.set_aspect('equal', 'box')
        for shape in self.shapes:
            shape.paint(ax)
        plt.xlim(0, 30)
        plt.ylim(0, 30)
        plt.show()

# ============================
# Demonstration of Usage
# ============================

if __name__ == "__main__":
    # Create the initial drawing
    drawing = RasterDrawing()

    # Add shapes
    rect = Rectangle(10, 5, 5, 5)
    circ = Circle(3, 20, 20)
    tri = Triangle(6, 5, 7, 15, 5)

    drawing.add_shape(rect)
    drawing.add_shape(circ)
    drawing.add_shape(tri)

    # Paint the initial drawing
    print("Painting the initial drawing...")
    drawing.paint()

    # Save the drawing
    drawing.save("drawing.txt")
    print("Drawing saved to 'drawing.txt'.")

    # Load the drawing back
    loaded_drawing = RasterDrawing.load("drawing.txt")
    print("Loaded the drawing from 'drawing.txt'.")

    # Paint the loaded drawing
    print("Painting the loaded drawing...")
    loaded_drawing.paint()
```
